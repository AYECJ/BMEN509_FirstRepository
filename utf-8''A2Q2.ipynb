```python
#library imports
#connor Mullen
#10131211

#image 1 
import os
import math
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from scipy import signal, ndimage

# Setup Jupyter notebook specifics
%matplotlib inline

# Numpy printing help
np.set_printoptions(precision=2)


#func1 = 20*math.sin(2*math.pi*f0*x) + 20*math.sin(2*math.pi*f0*y)
#func2 = 2*math.exp((-1)*(((x-5)**2/2)+((y-5)**2)/2))
#func3 = 20*math.sin(2*math.pi*f0*x)
#np.piecewise(func4, [-10 < x < 10, x >= 10, x <= -10], [1, 0, 0])
#np.piecewise(func4, [-10 < y < 10, y >= 10, y <= -10], [1, 0, 0])
#for x in range(-20, 20):
#    for y in range(-20, 20):
#        if (y > -10) and (y < 10) and (x > -10) and (x < 10):
#            func4 = 1
#        else:
#            func4 = 0
            
img1 = np.zeros((20,20))
for x in range(img1.shape[0]):
    for y in range(img1.shape[1]):
        img1[x,y] = 2*math.exp((-1)*(((x-5)**2/2)+((y-5)**2)/2))
        
sizeOfImage1 = img1.shape[0] * img1.shape[1]
freq1 = np.array(img1)
fourier1 = np.fft.fft(freq1)
fourier1abs = abs(fourier1)*2 / sizeOfImage1 * 2 / sizeOfImage1


plt.subplots(1, 3, figsize=(25,10))
plt.subplot(1,3,1)
plt.imshow(img1, cmap ='gray')
plt.subplot(1,3,2)
plt.hist(img1.ravel(), bins = 256, color = "orange")
plt.xlabel('Intensities')
plt.ylabel('Count')
plt.subplot(1,3,3)
plt.plot(fourier1abs)
plt.show()
        
            
    

#plt.hist(func1, bins = range(-30, 30), color = "red")
#plt.hist(func2, bins = range(-30, 30), color = "black")
#plt.hist(func3, bins = range(-30, 30), color = "blue")
```

```python
#image 2 code
f0 = 2
img2 = np.zeros((20,20))
for x in range(img2.shape[0]):
    for y in range(img2.shape[1]):
        img2[x,y] = 20*math.sin(2*math.pi*f0*x) + 20*math.sin(2*math.pi*f0*y)
        
        
sizeOfImage2 = img2.shape[0] * img2.shape[1]
freq2 = np.array(img2)
fourier2 = np.fft.fft(freq2)
fourier2abs = abs(fourier2)*2 / sizeOfImage2 * 2 / sizeOfImage2


plt.subplots(1, 3, figsize=(25,10))
plt.subplot(1,3,1)
plt.imshow(img2, cmap ='gray')
plt.subplot(1,3,2)
plt.hist(img2.ravel(), bins = 256, color= "blue")
plt.xlabel('Intensities')
plt.ylabel('Count')
plt.subplot(1,3,3)
plt.plot(fourier2abs)
plt.show()
```

```python
#image 3 code
f0 = 2
img3 = np.zeros((20,20))
for x in range(img3.shape[0]):
    for y in range(img3.shape[1]):
        img3[x][y] = 20*math.sin(2*math.pi*f0*x)
        
        
sizeOfImage3 = img3.shape[0] * img3.shape[1]
freq3 = np.array(img3)
fourier3 = np.fft.fft(freq3)
fourier3abs = abs(fourier3)*2 / sizeOfImage3 * 2 / sizeOfImage3


plt.subplots(1, 3, figsize=(25,10))
plt.subplot(1,3,1)
plt.imshow(img3, cmap ='gray')
plt.subplot(1,3,2)
plt.hist(img3.ravel(), bins = 256, color = "red")
plt.xlabel('Intensities')
plt.ylabel('Count')
plt.subplot(1,3,3)
plt.plot(fourier3abs)
plt.show()
```
